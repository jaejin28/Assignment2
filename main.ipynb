```python
import pandas
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
%matplotlib inline
df=pandas.read_csv('C:/Users/김재진/OneDrive/Desktop/DATA.csv')
print(df)
date_list=[]
for i in range(301,332):
    date_list.append(i)
for i in range(401,431):
    date_list.append(i)

df_groupby_date1=df.groupby('year_de')['river_hd','addr_tp','type_ve','area_num'].mean()
df_groupby_date2=df.groupby('year_de')['hi_tp','len_tp','rsqty_tp'].mean()

pandas.DataFrame(df_groupby_date1).to_csv('C:/Users/김재진/OneDrive/Desktop/관측소별_기상관측_정보_평균 기후.csv')
pandas.DataFrame(df_groupby_date2).to_csv('C:/Users/김재진/OneDrive/Desktop/관측소별_기상관측_정보_개화 요건.csv')
pandas.DataFrame(df_groupby_date1)
s1=pandas.Series(df_groupby_date1['river_hd'], index=date_list)
plt.figure(figsize=(15,10))
s1.plot(kind="bar")
plt.show()
s2=pandas.Series(df_groupby_date1['addr_tp'], index=date_list)
plt.figure(figsize=(15,10))
s2.plot(kind="bar")
plt.show()
s3=pandas.Series(df_groupby_date1['type_ve'], index=date_list)
plt.figure(figsize=(15,10))
s3.plot(kind="bar")
plt.show()
s4=pandas.Series(df_groupby_date1['area_num'], index=date_list)
plt.figure(figsize=(15,10))
s4.plot(kind="bar")
plt.show()
s5=pandas.Series(df_groupby_date2['hi_tp'], index=date_list)
plt.figure(figsize=(15,10))
s5.plot(kind="bar")
plt.show()
s6=pandas.Series(df_groupby_date2['len_tp'], index=date_list)
plt.figure(figsize=(15,10))
s6.plot(kind="bar")
plt.show()
s7=pandas.Series(df_groupby_date2['rsqty_tp'], index=date_list)
plt.figure(figsize=(15,10))
s7.plot(kind="bar")
plt.show()
```
